# Exploratory Data Analysis

```python
# Removing spaces betwwen words and converting into lower case to avoid typing errors
df['Annual Family Income'] = df['Annual Family Income'].str.lower().replace(r"(\s+)", "")
df['Annual Family Income'].unique()
```

```python
# Removing 2 blank rows, added at the end of the dataset.
dff = df.iloc[1:1095]
```

```python
# Checking null values
dff.isnull().sum()
```

```python
# Make the plot
plt.figure(figsize=(12,6))
plt.title("Percentage of lectures watched by viewer's Annual Family Income ",fontsize=15)
labels=dff['Annual Family Income'].value_counts().index
plt.pie(dff['Annual Family Income'].value_counts(),labels=labels , shadow=True, autopct='%1.1f%%',startangle=180)
plt.plot()
```

Most of lectures viewer's family income is below 3 lacs.

```python
# Make the plot
plt.figure(figsize=(15,10))

plt.title("Percentage of Lectures watched from different locations",fontsize=20)
labels=dff['Where do you live now (for internal purposes)?'].value_counts().index
plt.pie(dff['Where do you live now (for internal purposes)?'].value_counts(),labels=labels, shadow=True, autopct='%1.1f%%',startangle=180)
plt.plot()
```

Viewers from Towns and Rural Villages watch less lectures in comparison to viewers belong to Tier 1 and Tier 2 Cities.

```python
# Make the plot
plt.figure(figsize=(12,6))
plt.title("Percentage of watched Subject Area",fontsize=15)
labels=dff['What is your Subject Area?'].value_counts().index
plt.pie(dff['What is your Subject Area?'].value_counts(),labels=labels , shadow=True, autopct='%1.1f%%',startangle=180)
plt.plot()
```

Less watched  subject area is Others(BPharm, Agri, etc.)

```python
# Make the plot
plt.figure(figsize=(12,6))
plt.title("Percentage of Viwers from different Assigned categories",fontsize=15)
labels=dff['Assigned Category'].value_counts().index
plt.pie(dff['Assigned Category'].value_counts(),labels=labels , shadow=True, autopct='%1.1f%%',startangle=180)
plt.plot()
```

Most of the lectures are watched by Scholared.

```python
# Make the plot
plt.figure(figsize=(12,6))
plt.title("Percentage of viewers hear about us from different sources",fontsize=15)
labels=dff['How did you hear about us ?'].value_counts().index
plt.pie(dff['How did you hear about us ?'].value_counts(),labels=labels , shadow=True, autopct='%1.1f%%',startangle=180)
plt.plot()
```

Most of the viewers hear about us from Other sources.

```python
df12 = dff[['What is your Subject Area?', 'SUK Live - Total WatchTime   (Hour) ', 'MasterClass - Total WatchTime (Hours) ', 'Recorded Videos -Total WatchTime  (Hours) ']]
#df12
```
